```python
# Create data from an Equation
# E = mc^2

import numpy as np
import matplotlib.pyplot as plt
```

```python
m = np.linspace(0,10,11)
c = 299792458

E = m * (c**2)
```

```python
E
```

```python
# Now plotting E = mc^2

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(m,E, color="red",lw=4)
ax.set_ylabel("Energy in Joules")
ax.set_xlabel("Mass in grams")
ax.set_title("E=mc^2")

ax.set_xlim(0,10)
```

```python
# Now plotting E = mc^2 on log scale

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(m,E, color="red",lw=4)
ax.set_ylabel("Energy in Joules")
ax.set_xlabel("Mass in grams")
ax.set_title("E=mc^2")

ax.set_xlim(0,10)

ax.set_yscale('log')
ax.grid(which="both",axis="y")
```

```python
# Creating plots from Data Points

labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
#Plot Both cureves on the figure

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(labels,july16_2007,label="July 16, 2007")
ax.plot(labels,july16_2020, label="July 16, 2020")

ax.legend(loc=(1.1,0.5))

```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

#top plot
ax[0].plot(labels,july16_2007)
ax[0].set_title("July 16th 2007")

#bottom plot
ax[1].plot(labels,july16_2020)
ax[1].set_title("July 16th 2020")

plt.tight_layout()
```

```python
# Twin Axis Problem

fig, ax1 = plt.subplots(figsize=(10,5))

ax1.plot(labels,july16_2007)
ax1.set_ylabel("2007", color="blue")

ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(4)

for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(labels,july16_2020, color="red")
ax2.set_ylabel("2020", color="red")

ax2.spines['right'].set_color('red')
ax2.spines['right'].set_linewidth(4)

for label in ax2.get_yticklabels():
    label.set_color('red')

plt.tight_layout()
```
